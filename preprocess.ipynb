```python
import vcf
import csv
import gzip

# Path to the input VCF file
vcf_gz_file = '/Users/itscclemon/Desktop/CSE 284/1000G_chr1_pruned.vcf.gz'

# Path to the output CSV file
csv_file = '/Users/itscclemon/Desktop/CSE 284/1000G_chr1_pruned.csv'

# Open the compressed VCF file for reading
with gzip.open(vcf_gz_file, 'rt') as vcf_file:
    # Open the VCF file for reading using PyVCF
    vcf_reader = vcf.Reader(vcf_file)
    
    # Extract ID and FORMAT column headers
    headers = ['ID', 'FORMAT']
    
    # Create a list to store data rows
    rows = []
    
    # Iterate over each variant record in the VCF file
    for record in vcf_reader:
        # Extract ID and FORMAT values for the current variant
        id_value = record.ID
        format_value = record.FORMAT
        
        # Append ID and FORMAT values as a row to the list
        rows.append([id_value, format_value])
    
    # Write data to CSV file
    with open(csv_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        # Write headers
        writer.writerow(headers)
        # Write data rows
        writer.writerows(rows)

print("Conversion completed successfully!")
```

```python
import os
import vcf
import pandas as pd


def process_vcf(vcf_filename, output_csv_filename):
    try:
        vcf_reader = vcf.Reader(filename=vcf_filename)

        snp_ids = []
        individual_ids = []
        genotype_data = []

        for record in vcf_reader:
            snp_ids.append(record.ID)
            individual_ids.append([sample.sample for sample in record.samples])
            genotype_data.append([call.gt_bases for call in record.samples])

        genotype_df = pd.DataFrame(genotype_data, columns=individual_ids, index=snp_ids)

        genotype_df = genotype_df.T

        genotype_df.to_csv(output_csv_filename)
        print(f"Saved CSV file: {output_csv_filename}")

    except Exception as e:
        print(f"Error processing file {vcf_filename}: {e}")

vcf_directory = '/Users/itscclemon/Desktop/CSE 284'

output_directory = '/Users/itscclemon/Desktop/CSE 284'

for filename in os.listdir(vcf_directory):
    if filename.endswith('.vcf.gz'):
        vcf_filename = os.path.join(vcf_directory, filename)
        chromosome = filename.split('_')[1]  # Extract chromosome number from filename
        output_csv_filename = os.path.join(output_directory, f'chr{chromosome}_genotype_data.csv')
        process_vcf(vcf_filename, output_csv_filename)
```
