```python
train_r2_ENET_EUR = [enet_EUR[gene_id]['r2_train'] for gene_id in enet_EUR]
test_r2_ENET_EUR = [enet_EUR[gene_id]['r2_test'] for gene_id in enet_EUR]

train_r2_ENET_YRI = [enet_YRI[gene_id]['r2_train'] for gene_id in enet_YRI]
test_r2_ENET_YRI = [enet_YRI[gene_id]['r2_test'] for gene_id in enet_YRI]

train_r2_LASSO_EUR = [lasso_EUR[gene_id]['r2_train'] for gene_id in lasso_EUR]
test_r2_LASSO_EUR = [lasso_EUR[gene_id]['r2_test'] for gene_id in lasso_EUR]

train_r2_LASSO_YRI = [lasso_YRI[gene_id]['r2_train'] for gene_id in lasso_YRI]
test_r2_LASSO_YRI = [lasso_YRI[gene_id]['r2_test'] for gene_id in lasso_YRI]

train_r2_MARGINAL_EUR = [marginal_EUR[gene_id]['r2_train'] for gene_id in marginal_EUR]
test_r2_MARGINAL_EUR = [marginal_EUR[gene_id]['r2_test'] for gene_id in marginal_EUR]

train_r2_MARGINAL_YRI = [marginal_YRI[gene_id]['r2_train'] for gene_id in marginal_YRI]
test_r2_MARGINAL_YRI = [marginal_YRI[gene_id]['r2_test'] for gene_id in marginal_YRI]
```

```python
# density plot for all the r2 values
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
sns.kdeplot(train_r2_ENET_EUR, label='ENET_EUR_train', color='#0317fc')
sns.kdeplot(test_r2_ENET_EUR, label='ENET_EUR_test', color='#03fcfc')
sns.kdeplot(train_r2_ENET_YRI, label='ENET_YRI_train', color='#fc8c03')
sns.kdeplot(test_r2_ENET_YRI, label='ENET_YRI_test', color='#fcd703')
plt.xlabel('R2')
plt.ylabel('Density')
plt.title('R2 Density Plot')
plt.legend()
plt.show()
```
